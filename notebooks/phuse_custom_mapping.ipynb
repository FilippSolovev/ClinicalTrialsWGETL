## PHUSE_DM_RACE

```
%%sql

SELECT race     AS source_code, 
       race     AS description, 
       Count(*) AS cnt 
FROM   src.dm 
WHERE  race IS NOT NULL 
GROUP  BY race;
```

```
phuse_dm_race = %sql SELECT race     AS source_code, \
                            race     AS description, \
                            Count(*) AS cnt \
                            FROM   src.dm \
                            WHERE  race IS NOT NULL \
                            GROUP  BY race;
phuse_dm_race.to_csv('phuse_dm_race.csv', index=False)
files.download('phuse_dm_race.csv')
```
